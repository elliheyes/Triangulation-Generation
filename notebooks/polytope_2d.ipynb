```python
import numpy as np
import matplotlib.pyplot as plt


import sys
sys.path.append("..")
from build.py_polytope import Polytope

%matplotlib inline
```

```python
# Define polytope
points = [[1,0], [0,1], [-1,-1]]


# Note, all lattice points need to be manually specified for now.
p = Polytope(points, list(np.random.random(len(points))))
```

```python
# Replace vertex at infinity by the origin
facets = list(map(lambda x : list(map(lambda y: y if len(y) > 0 else [0.0,0.0], x)), p.getFacets()))
```

```python
# Plot the facets
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot()

colors = ['r','g','b', 'm']

for i in range(len(facets)):
    for j in range(len(facets[i])):
        for k in range(len(facets[i])):
            if j == k: continue
            facet_j = lambda j: facets[i][j % len(facets[i])]
            ax.plot(
                [facet_j(j)[0], facet_j(k)[0]],
                [facet_j(j)[1], facet_j(k)[1]], c='k', linewidth='3')


ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
```
